```python
import plotly.graph_objs as go
import numpy as np

# Generate random data points in 3D space
num_points = 10
x = np.random.randn(num_points)
y = np.random.randn(num_points)
z = np.random.randn(num_points)

# Human-meaningful labels for the cities
city_labels = ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix",
               "Philadelphia", "San Antonio", "San Diego", "Dallas", "San Jose"]

# Create a scatter plot for the data points with city labels
scatter = go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=10,
        color='blue',
    ),
    name='Data Points'
)


# Create the layout
layout = go.Layout(
    title='3D Embedding Space with Vector Line',
    scene=dict(
        xaxis=dict(title='X'),
        yaxis=dict(title='Y'),
        zaxis=dict(title='Z'),
    )
)

# Create the figure
fig = go.Figure(data=[scatter], layout=layout)

# Add annotations for each city
for i in range(num_points):
    fig.add_trace(
        go.Scatter3d(
            x=[x[i]],
            y=[y[i]],
            z=[z[i]],
            mode='text',
            text=[city_labels[i]],
            textposition='bottom center',
            showlegend=False
        )
    )

# Display the plot
fig.show()

```

```python
import plotly.graph_objects as go
import numpy as np

# Generate random data points (vectors) in 3D space
words = ['cat', 'dog', 'king', 'queen', 'car', 'hot', 'cold', 'big', 'large', 'small', 'tiny', "soft", "green"]
num_points = len(words)
np.random.seed(42)
data_points = np.random.randn(num_points, 3)

# Generate a random query vector
query_vector = np.random.randn(3)

# Calculate distances between query vector and data points
distances = np.linalg.norm(data_points - query_vector, axis=1)

# Find the index of the nearest neighbor
nearest_neighbor_index = np.argmin(distances)
nearest_neighbor_distance = distances[nearest_neighbor_index]

# Define colors for the data points
colors = ['blue'] * num_points
colors[nearest_neighbor_index] = 'red'  # Highlight the nearest neighbor

# Create a scatter plot for the data points
scatter = go.Scatter3d(
    x=data_points[:, 0],
    y=data_points[:, 1],
    z=data_points[:, 2],
    mode='markers',
    marker=dict(
        size=5,
        color=colors,
        opacity=0.5
    ),
    name='Data Points'
)

# Create a scatter plot for the query vector
query_scatter = go.Scatter3d(
    x=[query_vector[0]],
    y=[query_vector[1]],
    z=[query_vector[2]],
    mode='markers',
    marker=dict(
        size=10,
        color='green',
        symbol='cross'
    ),
        text="Query",
    name='Query Vector'
)

# Create a scatter plot for the nearest neighbor
nearest_neighbor_scatter = go.Scatter3d(
    x=[data_points[nearest_neighbor_index, 0]],
    y=[data_points[nearest_neighbor_index, 1]],
    z=[data_points[nearest_neighbor_index, 2]],
    mode='markers',
    marker=dict(
        size=10,
        color='red',
        symbol='cross'
    ),
            text="closest match",
    name='Nearest Neighbor'
)

# Add labels for each vector
annotations = [
    dict(
        x=data_points[i, 0],
        y=data_points[i, 1],
        z=data_points[i, 2],
        text=words[i],
        xshift=10,
        yshift=10,
        font=dict(
            color="black",
            size=12
        ),
        showarrow=False,
    )
    for i in range(num_points)
]

# Create the layout
layout = go.Layout(
    title='Vector Search Visualization in 3D Space',
    scene=dict(
        xaxis=dict(title='X'),
        yaxis=dict(title='Y'),
        zaxis=dict(title='Z'),
    )
)

# Add all traces and annotations to the figure
fig = go.Figure(data=[scatter, query_scatter, nearest_neighbor_scatter], layout=layout)
fig.update_layout(scene=dict(annotations=annotations),                 width=1200, height=800,)

# Display the plot
fig.show()
```

```python
# setup variables and helper functions

import chromadb #vector database
import ollama # LLM
from markdown import markdown
from bs4 import BeautifulSoup, NavigableString
import logging
import sys
from sentence_splitter import split_text_into_sentences

rootLogger=logging.getLogger()
# turn down logging for all other logs (e.g. http, url)
rootLogger.setLevel(logging.INFO)
rootLogger.handlers[0].setFormatter(logging.Formatter('[%(levelname)s] [%(name)s] %(message)s'))

log = logging.getLogger('cfb_script_import')
log.setLevel(logging.DEBUG)

#LLM_MODEL='llama3'
#LLM_MODEL='gemma:2b' #google deep mind trained model, small, inspired by google gemini
LLM_MODEL='tinyllama' # a smaller faster model
# cosine is a good distance algorithm for text retrieval. Can handle a bit of noise and fairly fast
DISTANCE_FUNC="cosine"
JOURNAL_COLLECTION=("journal_" + LLM_MODEL + "_" + DISTANCE_FUNC).replace(':', '_')
WIKI_COLLECTION=("wiki_page_" + LLM_MODEL + "_" + DISTANCE_FUNC).replace(':', '_')

# Initialize Chroma database client
chroma_client = chromadb.HttpClient(host="127.0.0.1", port=8000)

# Create the collection in the Chroma database
# these are like SQL tables

journal_collection = chroma_client.get_or_create_collection(JOURNAL_COLLECTION,
    # Valid options for hnsw:space are "l2", "ip, "or "cosine". The default is "l2" which is the squared L2
    metadata={"hnsw:space": DISTANCE_FUNC }
)

wiki_collection = chroma_client.get_or_create_collection(WIKI_COLLECTION,
    # Valid options for hnsw:space are "l2", "ip, "or "cosine". The default is "l2" which is the squared L2
    metadata={"hnsw:space": DISTANCE_FUNC }
)

def llm_get_embeddings(prompt:str):
    response=ollama.embeddings(model=LLM_MODEL, prompt=prompt)
    #log.debug(response)
    embedding_vector = response['embedding']
    #log.debug(embedding_vector)
    return embedding_vector

def llm_prompt(prompt:str):
    response=ollama.chat(model=LLM_MODEL, messages=[{ 'role':'user', 'content':prompt}])
    return response['message']['content']

def read_file(file_path:str):
    log.info("reading file " + file_path)
    try:
        with open(file_path, 'r') as file:
            content = file.read()
        return content
    except FileNotFoundError:
        print("File not found.")
        return None

def md_to_jsoup(text_md:str):
  text_html = markdown(text_md)
  return BeautifulSoup(text_html, "html.parser")

def split_text(text:str):
  return split_text_into_sentences(text=text, language='en')


# convert from a question into a statement
# e.g 'How do I do X?' to 'To do X' This might help with better matches
def convert_to_imperative(question:str):
    imperative=llm_prompt( f"""
      Convert the following single sentence from a question style to an imperative one, being as concise as possible, and using a neutral tone:

      {question}

      For example, to convert "How do I do X", respond only with  "To do X". Do not end with a full stop. Do not include the actual steps. Do not embelish the answer

      It it very important that you provide the final output without any additional comments or remarks, and that you do not attempt to answer the question
      """)
    log.debug("convert_to_imperative: question:'%s', imperative:'%s'", question, imperative)
    return imperative


def convert_to_past_tense(question:str):
    past_tense=llm_prompt( f"""
      Hello assistant. Revise the following text to convert it from a question to a past tense style, as would appear in a journal. Adhere strictly to these guidelines:

      1. Be as concise as possible, and using a neutral tone.
      2. Eliminate unclear text, such as excessive symbols or gibberish.
      3. Shorten text while preserving information.
      4. Remove standalone numbers or letters lacking context.
      5. Maintain clarity and information of the original text.
      6. Directly provide only the revised text.
      7. Do not answer the question or provide any hints as to how it might be resolved
      8. No pre-ample or friendly 'sure'

      {question}
      """)
    log.debug("convert_to_past_tense: question:'%s', past_tense:'%s'", question, past_tense)
    return past_tense
```

```python
# install and load the LLM model
# note that ollama can run multiple models at once. However they all take resources, so best to unload the ones you are not using
log.info("Installing Llama model '%s'", LLM_MODEL)
ollama.pull(LLM_MODEL)
log.info("pulled model")

models=ollama.list()['models']
for model in models:
  details=model['details']
  #log.info("Installed models: %s", model)
  log.info("Installed model: name=%s, size:%s, family=%s, parameter_size=%s", model['name'], model['size'], details['family'], details['parameter_size'])
```

```python
# insert data journal data

JOURNAL_FILES=['journal_april.md', 'journal_may.md', 'journal_june.md']

def main():
  insert_data()\


def insert_data():
  for file in JOURNAL_FILES:
    insert_journal_file(file)


def insert_journal_file(file:str):
    soup = md_to_jsoup(read_file(file))

    # Extract entries from HTML
    current_month = ""
    entries = []

    for element in soup.find_all(['h1', 'h2', 'li']):
        if element.name == 'h1':
            if entries:
                insert_journal_entries(file, current_month, entries)
                entries = []
            current_month = element.text
        elif element.name == 'h2':
            if entries:
                insert_journal_entries(file, current_month, entries)
            entries = []
            current_date = element.text
        elif element.name == 'li':
            time_activity = element.text
            entries.append({
                "date": current_date,
                "activities": [time_activity]
            })

    # Insert remaining entries
    if entries:
        insert_journal_entries(file, current_month, entries)
    log.info("inserted journal file " + file)


# Function to insert diary entries into Chromas
def insert_journal_entries(source:str, month:str, entries:list):
    for entry in entries:
        id = entry["date"]
        date = entry["date"]

        activities = entry["activities"]
        # Convert activities to a single string (you might want to use a more sophisticated encoding)
        activities_str = " ".join(activities)
        log.debug("activities=%s", activities)
        # skip existing entries
        exists = len(journal_collection.get(ids=[id])) > 0
        if exists:
            log.debug('exists, skipping %s,  month %s, date %s', source, month, date)
            continue
        log.debug('inserting %s,  month %s, date %s', source, month, date)

        # convert to vector embeddings
        vector=llm_get_embeddings(activities_str)
        # Insert into Chroma
        journal_collection.add(
            documents=[activities_str],
            embeddings=[vector],
            metadatas=[{
                "source": source,
                "page": source,
                "month": month,
                "date": date
            }],
            ids=[ id ]
        )

main()
```

```python
# insert wiki data

WIKI_FILES=['wiki_1.md', 'wiki_2.md', 'wiki_3.md', 'wiki_4.md', ]

def main():
  insert_data()


def insert_data():
  for file in WIKI_FILES:
    insert_wiki_file(file)

  log.info("inserted all data!")


def insert_wiki_file(file:str):
    chunks = parse_wiki_markdown_into_chunks(read_file(file))
    for chunk in chunks:
        title = chunk["title"]
        content = chunk["content"]
        id=file + ":title"
        exists = len(wiki_collection.get(ids=[id])) > 0
        exists = False
        #log.debug("chunk %s", content)
        if exists:
            log.debug('exists, skipping %s,  title %s', file, title)
            continue
        log.debug('inserting chunk: source=%s, title=%s, content=%s', file, title, content)
        vector = llm_get_embeddings(title + " " + content)
        wiki_collection.add(
            documents = [content],
            embeddings = [vector],
            metadatas = [{
                "source": file,
                "title": title
            }],
          ids=[ id ]
        )
    log.info("todo: inserted wiki file " + file)


def parse_wiki_markdown_into_chunks(content_md:str):
  """
  Parses markdown content into chunks, chunked per heading

  Args:
    content_md: The markdown content to parse.

  Returns:
    A list of chunks.
  """

  soup = md_to_jsoup(content_md)

  # Find all headings in the markdown content
  headings = soup.find_all(["h1", "h2", "h3", "h4", "h5", "h6"])
  chunks = []
  # Iterate over the headings and create chunks
  for heading in headings:
    log.debug("heading: " + heading.text)
    # Create a new chunk with the heading as the title
    chunk = {"title": heading.text, "content": ""}
    chunks.append(chunk)

    # walk through all siblings of the heading that are not headings to append
    # the text
    sibling = heading.find_next_sibling()
    while sibling:
      is_heading=sibling.name.startswith("h")
      if is_heading:
        # go handle next heading
        break
      # todo: split into sentences, to have smaller chunks
      if not isinstance(sibling, NavigableString):
        chunk["content"] += str(sibling.text)
      sibling = sibling.find_next_sibling()
  return chunks


main()
```
